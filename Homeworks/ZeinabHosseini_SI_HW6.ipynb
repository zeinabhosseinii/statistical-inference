#### **e)**

```r
# Compute the fits
# Kernel Smoothing Fit
ksmooth_fit <- ksmooth(df$age, df$lweight, kernel = "normal", bandwidth = 5)

# Smoothing Spline Fit
spline_fit <- smooth.spline(df$age, df$lweight)

# Create a base ggplot with scatter plot
p <- ggplot(df, aes(x = age, y = lweight)) +
  geom_point(alpha = 0.5, color = "blue") +  # Scatter plot with transparency
  labs(title = "Log Weight vs Age with Multiple Fits",
       x = "Age",
       y = "Log Weight (lweight)") +
  theme_minimal()

# Add Kernel Smoothing Fit
p <- p + geom_line(aes(x = ksmooth_fit$x, y = ksmooth_fit$y),
                   data = data.frame(age = ksmooth_fit$x, lweight = ksmooth_fit$y),
                   color = "purple", size = 1, linetype = "dashed")

# Add Smoothing Spline Fit
p <- p + geom_line(aes(x = spline_fit$x, y = spline_fit$y),
                   data = data.frame(age = spline_fit$x, lweight = spline_fit$y),
                   color = "orange", size = 1)

# Add LOESS Fit with 95% Confidence Band
p <- p + geom_smooth(method = "loess", se = TRUE, color = "deeppink", fill = "gray")

# Display the plot
print(p)
```

It is clear that the three fitted lines are relatively similar and closely aligned with each other, particularly in the central region where most of the data points are concentrated.

However, there are some differences between the fits:

$1.$ The kernel fit seems to be more influenced by the outliers.

$2.$ The smoothing spline fit also shows some non-linearity, but to a lesser extent than the kernel fit, likely due to the smoothing property of the method.

$3.$ The linear fit, appears as a straight line and is less affected by the outliers compared to the kernel and smoothing spline fits.

Overall, while there are minor differences among the three fits, they all provide a reasonably good approximation to the data.
